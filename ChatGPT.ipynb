<h3>ChatGPT API import and prediction</h3>

```python
from openai import OpenAI
import random
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

client = OpenAI(api_key = "key")

comments = pd.read_csv("/Users/mazichang/Desktop/Cyberviolence/Project/CommentsForPretrainedGPT.csv")

# Only need the comments for ChatGPT to process
contents = []
for index, row in comments.iterrows():
    contents.append(row['content'])

output = []
# If the API disconnected, can continue starting from the value of position
position = 1

for content in contents:
    
    # Track position
    if position%100 == 0:
        print(position)
    position = position+1
    
    # Tell ChatGPT to append either '1' or '0' to the output array
    task = f"Please indicate whether the following text is hate speech or not. Responsd with 1 for yes or 0 for no.\n{content}"
    completion = client.chat.completions.create(
      model="gpt-4o",
      messages=[
        {"role": "system", "content": "You are an internet moderator in detecting online hate speech."},
        {"role": "user", "content": task}
      ]
    )
    output.append(completion.choices[0].message.content)
```

<h3>Check if ChatGPT produced any erroneous response</h3>

```python
set(output)
# Sometimes it responds 'Yes', which is equivalent to a '1'
```

<h3>Clean the output</h3>

```python
output_cleaned = []
for x in output:
    if x.startswith('1') or x.startswith('Yes'):
        output_cleaned.append(1)
    elif x.startswith('0'):
        output_cleaned.append(0)
    else:
        output_cleaned.append(random.randint(0, 1))
```

<h3>Sanity check: Ensure that the cleaned output has the same length as the input comments

```python
assert len(output_cleaned) == len(output) == len(comments)
```

<h3>Save the cleaned output</h3>

```python
comments['gpt_predicted'] = output_cleaned
comments
comments.to_csv("CommentsCleanedWithGPT.csv", index=False)
```

<h3>Locate the testing set and calculate the four metrics</h3>

```python
train, test = train_test_split(comments, test_size=0.1, random_state=31)
testing_labels_range = test["label"]
```

```python
predicted_labels = test["gpt_predicted"]
accuracy = accuracy_score(testing_labels_range, predicted_labels)
recall = recall_score(testing_labels_range, predicted_labels)
precision = precision_score(testing_labels_range, predicted_labels)
f1 = f1_score(testing_labels_range, predicted_labels)
print("Accuracy", accuracy)
print("Recall", recall)
print("Precision", precision)
print("F1", f1)
```
